```python
import torch
import matplotlib.pyplot as plta

trainloss = []
valloss = []
# continue_from = ['models/Model1', 'models/Model2', 'models/Model3', 'models/Model4', 'models/Model5']
continue_from = ['models/Model{}'.format(x) for x in range(4)] 

check_per_epoch = -1

for f in continue_from:
    saved_model = torch.load(f, map_location=lambda storage, location: storage)
    if check_per_epoch == -1:
        check_per_epoch = len(saved_model['val_loss'])
    trainloss += saved_model['loss'][-check_per_epoch:]
    valloss += saved_model['val_loss'][-check_per_epoch:]
    print(trainloss[-1])
```

```python
plta.plot(trainloss, label='train')
plta.plot(valloss, label='validation')
plta.xlabel('steps (x2400)')
plta.ylabel('loss')
plta.title('{} epochs'.format(len(continue_from)))

plta.legend(loc='upper right')
plta.show()
```
